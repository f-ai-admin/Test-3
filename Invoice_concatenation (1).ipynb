## order details

Combining both the amazon and flipkart excel sheets with invoice details.

Merging the columns with similar data into a common excel file within a common column name.

example: 'Ship Address' in flipkart and 'Billing Address' in amazon --> Billing Address in final excel sheet

```python
import pandas as pd
```

```python
amazon=pd.read_excel('/content/Amazon_invoice_final.xlsx',sheet_name='order details')
```

```python
amazon.head()
```

```python
flipkart=pd.read_excel('/content/Flipkart_invoice_final.xlsx',sheet_name='order details')
```

```python
flipkart.head()
```

```python
# Check if 'Sold By' column exists in both DataFrames
if 'Sold By' in amazon.columns and 'Sold By' in flipkart.columns:
    column_head = 'Sold By'

    # Create a new DataFrame with only the 'Sold By' column
    df_sold = pd.concat([amazon[column_head].dropna(), flipkart[column_head].dropna()], ignore_index=True)
    df_sold = df_sold.rename('Sold By')
else:
    print("Column 'Sold By' does not exist")
df_sold
```

```python
# Check if 'Sold to' column exists in both DataFrames
if 'Billing Address' in amazon.columns and 'Sold to' in flipkart.columns:
    # Create a new DataFrame with only the 'Sold to' column
    df_sold_to = pd.concat([amazon['Billing Address'].dropna(), flipkart['Sold to'].dropna()], ignore_index=True)
    df_sold_to = df_sold_to.rename('Sold to')
else:
    print("Column 'Sold to' does not exist")
df_sold_to
```

```python
# Check if 'Address' column exists in both DataFrames
if 'Billing Address' in amazon.columns and 'Ship Address' in flipkart.columns:
    # Create a new DataFrame with only the 'Billing Address' column
    df_add = pd.concat([amazon['Billing Address'].dropna(), flipkart['Ship Address'].dropna()], ignore_index=True)
    df_add = df_add.rename('Billing Address')
else:
    print("Column 'Address' does not exist")
df_add
```

```python
# Check if 'Order ID' column exists in both DataFrames
if 'Order Number' in amazon.columns and 'Order ID' in flipkart.columns:
    # Create a new DataFrame with only the 'Order ID' column
    df_od_id = pd.concat([amazon['Order Number'].dropna(), flipkart['Order ID'].dropna()], ignore_index=True)
    df_od_id = df_od_id.rename('Order ID')
else:
    print("Column 'Order ID' does not exist")
df_od_id
```

```python
# Check if 'Order ID' column exists in both DataFrames
if 'Order Date' in amazon.columns and 'Order Date' in flipkart.columns:
    # Create a new DataFrame with only the 'Order Date' column
    df_od_d = pd.concat([amazon['Order Date'].dropna(), flipkart['Order Date'].dropna()], ignore_index=True)
    df_od_d = df_od_d.rename('Order Date')
else:
    print("Column 'Order Date' does not exist")
df_od_d
```

```python
# Check if 'Invoice Number' column exists in both DataFrames
if 'Invoice Number' in amazon.columns and 'Invoice Number' in flipkart.columns:
    # Create a new DataFrame with only the 'Invoice Number' column
    df_in_id = pd.concat([amazon['Invoice Number'].dropna(), flipkart['Invoice Number'].dropna()], ignore_index=True)
    df_in_id = df_in_id.rename('Invoice ID')
else:
    print("Column 'Invoice ID' does not exist")
df_in_id
```

```python
# Check if 'Invoice ID' column exists in both DataFrames
if 'Order Date' in amazon.columns and 'Invoice Date' in flipkart.columns:
    # Create a new DataFrame with only the 'Invoice Date' column
    df_in_d = pd.concat([amazon['Invoice Date'].dropna(), flipkart['Invoice Date'].dropna()], ignore_index=True)
    df_in_d = df_in_d.rename('Invoice Date')
else:
    print("Column 'Invoice Date' does not exist")
df_in_d
```

```python
# Check if 'Payment method' column exists in both DataFrames
if 'Payment method' in amazon.columns and 'Payment method' in flipkart.columns:
    # Create a new DataFrame with only the 'Payment method' column
    df_pay_met = pd.concat([amazon['Payment method'].dropna(), flipkart['Payment method'].dropna()], ignore_index=True)
    df_pay_met = df_pay_met.rename('Payment method')
else:
    print("Column 'Payment method' does not exist")
```

```python
order_details = pd.merge(df_sold, df_sold_to, how='outer', left_index=True, right_index=True)
order_details = pd.merge(order_details, df_add, how='outer', left_index=True, right_index=True)
order_details = pd.merge(order_details, df_od_id, how='outer', left_index=True, right_index=True)
order_details = pd.merge(order_details, df_in_id, how='outer', left_index=True, right_index=True)
order_details = pd.merge(order_details, df_od_d, how='outer', left_index=True, right_index=True)
order_details = pd.merge(order_details, df_in_d, how='outer', left_index=True, right_index=True)
order_details = pd.merge(order_details, df_pay_met, how='outer', left_index=True, right_index=True)
```

```python
order_details
```

## Product Summary

```python
amazon=pd.read_excel('/content/Amazon_invoice_final.xlsx',sheet_name='Invoice 1')
```

```python
amazon.head()
```

```python
flipkart1=pd.read_excel('/content/Flipkart_invoice_final.xlsx',sheet_name='Invoice 1')
flipkart2=pd.read_excel('/content/Flipkart_invoice_final.xlsx',sheet_name='Invoice 2')
flipkart3=pd.read_excel('/content/Flipkart_invoice_final.xlsx',sheet_name='Invoice 3')
flipkart4=pd.read_excel('/content/Flipkart_invoice_final.xlsx',sheet_name='Invoice 4')
```

```python
# Create a list of Flipkart DataFrames
flipkart_dfs = [flipkart2, flipkart3, flipkart4,flipkart1]

# Concatenate the DataFrames along the rows (axis=0)
combined_flipkart = pd.concat(flipkart_dfs, ignore_index=True)

# Display the combined DataFrame
combined_flipkart
```

```python
# Check if 'Description' column exists in both DataFrames
if 'Description' in amazon.columns and 'Item' in combined_flipkart.columns:
    # Create a new DataFrame with only the 'Description' column
    df_des = pd.concat([amazon['Description'].dropna(), combined_flipkart['Item'].dropna()], ignore_index=True)
    df_des = df_des.rename('Description')
else:
    print("Column 'Description' does not exist")
```

```python
df_des
```

```python
# Check if 'Quantity' column exists in both DataFrames
if 'Qty' in amazon.columns and 'Qty' in combined_flipkart.columns:
    # Create a new DataFrame with only the 'Qty' column
    df_q = pd.concat([amazon['Qty'].dropna(), combined_flipkart['Qty'].dropna()], ignore_index=True)
    df_q = df_q.rename('Quantity')
else:
    print("Column 'Quantity' does not exist")
df_q
```

```python
# Check if 'Tax Amount' column exists in both DataFrames
if 'Tax\nAmount' in amazon.columns and 'Tax Amount' in combined_flipkart.columns:
    # Create a new DataFrame with only the 'Tax' column
    df_tax = pd.concat([amazon['Tax\nAmount'].dropna(), combined_flipkart['Tax Amount'].dropna()], ignore_index=True)
    df_tax = df_tax.rename('Tax Amount')
else:
    print("Column 'Tax Amount' does not exist")
```

```python
df_tax
```

```python
# Check if 'Total Amount' column exists in both DataFrames
if 'Total\nAmount' in amazon.columns and 'Total' in combined_flipkart.columns:
    # Create a new DataFrame with only the 'Total' column
    df_tot = pd.concat([amazon['Total\nAmount'].dropna(),combined_flipkart['Total'].dropna()], ignore_index=True)
    df_tot = df_tot.rename('Total Amount')
else:
    print("Column 'Total Amount' does not exist")
```

```python
df_tot
```

```python
pro_details = pd.merge(df_des, df_q, how='outer', left_index=True, right_index=True)
pro_details = pd.merge(pro_details, df_tax, how='outer', left_index=True, right_index=True)
pro_details = pd.merge(pro_details, df_tot, how='outer', left_index=True, right_index=True)
pro_details.head(10)
```
