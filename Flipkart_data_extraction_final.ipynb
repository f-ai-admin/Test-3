```python
import pdfplumber


pdf_path = 'flipkart.pdf'

# Open the PDF file
with pdfplumber.open(pdf_path) as pdf:
    # Initialize an empty string to store extracted text from all pages
    all_text = ""

    # Loop through all pages
    for page_number in range(len(pdf.pages)):
        # Extract text from the current page
        page = pdf.pages[page_number]
        text = page.extract_text()

        # Append the text from the current page to the overall text
        all_text += text

        # print text from each page
        print(f"Text from Page {page_number + 1}:\n{text}\n\n\n")

# Print the combined text from all pages
print("Combined Text from All Pages:\n", all_text)
```

```python
import pandas as pd
import re

sold_by_values = []

    # Loop through each page
for page_number in range(len(pdf.pages)):
        # Extract text from the current page
        text = pdf.pages[page_number].extract_text()

        # Find all occurrences of 'Sold By' on the current page
        start_positions = [m.start() for m in re.finditer('Sold By', text)]

        # Loop through each occurrence on the current page
        for start in start_positions:
            end = text.find(',', start)
            sold = text[start:end].replace('\n', ' ').split(':', 1)[1].strip()
            sold_by_values.append(sold)

    # Check if any 'Sold By' values were found
if sold_by_values:
        # Create a DataFrame with all 'Sold By' values
        df1 = pd.DataFrame({'Sold By': sold_by_values})
        print(df1)
else:
        print("No 'Sold By' found in the text.")
```

## Order Summary

```python
import pandas as pd
import re

items = []

# Loop through each page
for page_number in range(len(pdf.pages)):
    # Extract text from the current page
    text = pdf.pages[page_number].extract_text()

    # Find start and end positions for 'FOOD ITEMS' and 'Summary' on the current page
    start = text.find('FOOD ITEMS')
    end = text.find('Summary', start)

    if start != -1 and end != -1:
        food_items_text = text[start:end]

        # Split the text after every comma and extract 'Sold By' values
        for items_match in re.finditer(r'([^)]*)', food_items_text):
            item = items_match.group(1).strip()

            # Remove 'FOOD ITEMS' only from the row
            item = item.replace('FOOD ITEMS', '').strip()
            items.append(item)

# Check if any 'Items' values were found
if items:
    # Create a DataFrame with all 'Items' values
    df10 = pd.DataFrame({'Item HSN(Tax%) Qty MRP Savings Total': items})
    print(df10)
else:
    print("No 'Items' found in the text.")
```

# Extracting the product description in the trapped tabel using camelot

```python
import camelot

pdf_path = 'flipkart.pdf'

tables = camelot.read_pdf(pdf_path, flavor='stream', pages='all')

# Filter tables containing 'Summary'
filtered_tables = [table for table in tables if 'Summary' in ' '.join(map(str, table.df.values.flatten()))]

# Display remaining tables
for i, table in enumerate(filtered_tables):
    print(f"Table {i + 1}:\n{table.df}\n")
```

```python
start= 'S.'
end= 'Summary'

extracted_tables = []


for i, table in enumerate(filtered_tables):

    # Finding the index of the row containing the start keyword
    start_index = table.df.index[table.df.apply(lambda row: start in ' '.join(map(str, row)), axis=1)].tolist()[0]

    # Finding the index of the row containing the end keyword
    end_index = table.df.index[table.df.apply(lambda row: end in ' '.join(map(str, row)), axis=1)].tolist()[0]

   #Extracts rows between the start and the end
    extracted_row = table.df.loc[start_index:end_index]

    # Make the first row as column headers
    extracted_row.columns = extracted_row.iloc[0]

    # Append the extracted DataFrame to the list
    extracted_tables.append(extracted_row.iloc[1:].reset_index(drop=True))

    # Display the extracted row
    print(f"Table {i + 1}, Extracted Row:\n{extracted_row}\n")

# Display the list of extracted DataFrames
for i, extracted_table in enumerate(extracted_tables):
    print(f"Extracted Table {i + 1}:\n{extracted_table}\n")
```
