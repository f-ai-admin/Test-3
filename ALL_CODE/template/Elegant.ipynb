```python
import fitz
import pandas as pd
import re
import os
import fitz
from PIL import Image
import pytesseract
```

```python
pdf_path = 'ELE2.pdf'

text = ''
pdf_document = fitz.open(pdf_path)
count = pdf_document.page_count
for i in range(count):
    page = pdf_document[i]
    tx= page.get_text()
    text+=tx
# print(text)


pattern = re.compile(r'Guest Name :([^0-9]+)\s+(\d+)\s+([\d.]+)\s+([\d.]+)\s+Journey Start Date :(\d{2}-\d{2}-\d{4})\s+Journey end Date:(\d{2}-\d{2}-\d{4})\s+Invoice No :(\w+-\d+)')

# Find all matches
matches = pattern.findall(text)
k = {}
k['Guest Name']  = []
k['Quantity'] = []
k['Rate'] = []
k['Amount'] = []
k['Journey start Date'] = []
k['Journey end Date'] = []
k['Invoice No'] = []
 

for match in matches:
    Guest_Name =  match[0].strip()
    k['Guest Name'].append(Guest_Name)
    
    Quantity =  match[1]
    k['Quantity'].append(Quantity)
    
    Rate = match[2]
    k['Rate'].append(Rate)
    
    Amount = match[3]
    k['Amount'].append(Amount)
    
    Journey_start_Date =  match[4]
    k['Journey start Date'].append(Journey_start_Date)
    
    Journey_end_Date = match[5]
    k['Journey end Date'].append(Journey_end_Date)
    
    Invoice_No = match[6]
    k['Invoice No'].append(Invoice_No)
    

df = pd.DataFrame(k)
df.to_excel('ele_out2.xlsx',index=False)
```
